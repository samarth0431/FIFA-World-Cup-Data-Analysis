```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Goals Scored per Tournament (Matplotlib)
plt.figure(figsize=(10, 6))
sns.lineplot(data=world_cup_data, x='Year', y='GoalsScored', marker='o')
plt.title('Goals Scored per Tournament')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
plt.grid(True)
plt.show()

import streamlit as st

# Title and introduction
st.title('World Cup Analysis Dashboard')
st.markdown('Explore historical World Cup data and insights.')

# Sidebar for filters or selections (if needed)
st.sidebar.title('Filters')
selected_year = st.sidebar.slider('Select Year', min_value=1930, max_value=2018, step=4)

# Visualizations section
st.header('Goals Scored per Tournament')
plt.figure(figsize=(10, 6))
sns.lineplot(data=world_cup_data, x='Year', y='GoalsScored', marker='o')
plt.title('Goals Scored per Tournament')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
plt.grid(True)
st.pyplot()

# Add more visualizations as needed
```
